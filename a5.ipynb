```python
import numpy as np
import pandas as pd

df = pd.read_excel("Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")

df = df.replace({'t': 1, 'f': 0, True: 1, False: 0})

binary_cols = []
for col in df.columns:
    unique_vals = df[col].dropna().unique()
    if set(unique_vals).issubset({0, 1}):
        binary_cols.append(col)

binary_data = df[binary_cols].values

v1 = binary_data[0]
v2 = binary_data[1]

f11 = np.sum((v1 == 1) & (v2 == 1))
f01 = np.sum((v1 == 0) & (v2 == 1))
f10 = np.sum((v1 == 1) & (v2 == 0))
f00 = np.sum((v1 == 0) & (v2 == 0))

JC = f11 / (f11 + f01 + f10)

SMC = (f11 + f00) / (f11 + f01 + f10 + f00)

print("Binary attributes used:", len(binary_cols))
print("f11:", f11)
print("f01:", f01)
print("f10:", f10)
print("f00:", f00)
print("Jaccard Coefficient (JC):", JC)
print("Simple Matching Coefficient (SMC):", SMC)
```
