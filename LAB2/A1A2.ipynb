```python
import pandas as pd
import numpy as np

data = pd.read_excel("Lab Session Data.xlsx", sheet_name="Purchase data")

X = data[["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
Y = data["Payment (Rs)"].values

rank_X = np.linalg.matrix_rank(X)
X_pinv = np.linalg.pinv(X)
price_c = X_pinv @ Y

print("Rank of X:", rank_X)
print("Prices:", price_c)
```

```python
import pandas as pd
import numpy as np

data = pd.read_excel("Lab Session Data.xlsx", sheet_name="Purchase data")
X = data[["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
Y = data["Payment (Rs)"].values

customer_class = []

for payment in Y:
    if payment > 200:
        customer_class.append("RICH")
    else:
        customer_class.append("POOR")

data["Customer Class"] = customer_class

print("Feature matrix X:\n", X)
print("\nPayment vector Y:\n", Y)
print("\nCustomer Classification (RICH / POOR):\n")
print(data[["Customer", "Payment (Rs)", "Customer Class"]])
```
